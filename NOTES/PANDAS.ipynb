## DataFrame indexing + data selection

 Pandas uses df.iloc[] for integer-based indexing to select data by position:
 It indexes single values and multiple rows and columns via slicing
df.iloc[row_start : row_end : row_step, 
        col_start : col_end : col_step]

```python
# DataFrame indexing + data selection
# Integer based indexing is by df.iloc[]
# Select 6 rows, last 3 columns
bsrn.iloc[1434:1440,12:]
```

```python
## first 5 columns , every 40th row

bsrn.iloc[::40,:5]
```

### Row indexing

Rows of DataFrame can be accessed using df.loc, which "locates" rows based on their labels. The row is transformed into a Series, the coumn name as the index:

```python
## index a single row

bsrn.loc[1434]

one_row = bsrn.loc[0]
one_row
```

The column contains mixed data types-floats,integers, and an object in the first row
The stop value is inclusive

```python
bsrn.loc[1434:1440]
```
